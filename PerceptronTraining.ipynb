# Problem Statement

Teach model to predict on the following equation 

Equation: `y = 10x`

# Data Creation


```
x = [0, 1, 2, 3, 4, 5,..]

y = [0, 10, 20, 30, 40, 50,..]
```

````python
import numpy as np
x = np.array([i for i in range(100)])
y = np.array([i*10 for i in x])
````

# Split the Data in Training & Testing

````python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2, random_state=5)

xTrain = np.expand_dims(xTrain, axis=1)
xTest = np.expand_dims(xTest, axis=1)
````

# Import the Libraries for Training the Model

````python
import tensorflow as tf
from tensorflow.keras import *
from tensorflow.keras.layers import *
````

````python
model = Sequential([
                    Dense(units=1, input_shape=xTrain.shape[1:]) 
])
````

### Mean Absolute Error

$${\displaystyle \mathrm {MAE} ={\frac {\sum _{i=1}^{n}\left|y_{i}-x_{i}\right|}{n}}}
$$

where,

- $y_{i}$ denotes the true label (the labels you have)
- $x_{i}$ denotes the predicted labels that the model outputs 

````python
model.compile(optimizer='adam', loss='mae')
````

````python
numEpoch = 5000

model.fit(x=xTrain, y=yTrain, validation_data=(xTest, yTest), epochs=numEpoch)
````

````python
model.evaluate(xTest, yTest)
````
